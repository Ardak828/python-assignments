```python
def fullpoem():
    """ this new poem is taken from the HePo poetry """
    
    t = '   Robots and Puppets   '
    w = 'Shaima Al-Marzouqi'
    
    print(t.upper().strip())
    print ('\n')
    a = 'Robots ' * 3
    print(a)
    print(f'All I see is {a[:6].lower()}')
    
    print ('\n')
    b = ['Where', 'are', 'the', 'humans']
    li2b = ' '.join(b)
    print(li2b + '?')
    
    c = ['They', 'are', 'long', 'gone', 'my', 'friend']
    li2c = ' '.join(c)
    print(li2c)
    print('\n')
    
    d = 'Everyone'
    print(f"{d} is controlled by \"them\"")
    print(f'{d} is programmed the same')
    
    words = {'first': 'Everywhere',
             'second': ['I', 'people', 'things'],
             'third': ['turn', 'see', 'doing'],
             'fourth': ['different', 'exact', 'same'],
             'fifth': 'the'      
    }   
    print(words['first'], words['second'][0], words['third'][0], words['second'][0], words['third'][1], words['fourth'][0], words['second'][1], words['third'][2], words['fifth'], words['fourth'][1], words['fourth'][2], words['second'][2])
    
    print('\n')
    print("\"We are all the same\", he said")
    print("\"Why do you say it proudly?\", I asked")
    print("\"We like it\", he answered")
    print("\"You are programmed that way,\n programmed to think you like it.\" I replied")
    
    print('\n')
    print("They are manipulating you, don't you see? \nYou are all bunch of robots and puppets")
    
    print('\n')
    e = ['you', 'want', 'live', 'life']
    print(f'Be who {e[0].upper()} {e[1]} to be')
    print(f'{e[2].title()} the {e[3]} {e[0].upper()} {e[1]} to {e[2]}')
    
    print('\n')
    morewords = {
        '1': ['do', 'chose', 'be'],
        '2': ['to', 'the', 'for', 'they', 'or'],
        '3': ['yourself', 'a puppet', 'free']
    }
    print(morewords['1'][0].title(), e[0], e[1], morewords['2'][0], e[2], morewords['2'][1], e[3], morewords['2'][3], morewords['1'][1], morewords['2'][2], e[0] + '?')
    print(morewords['2'][4])
    print(morewords['1'][0].title(), e[0], e[1], morewords['2'][0], e[2], morewords['2'][1], e[3], morewords['2'][3], morewords['1'][1], morewords['2'][2], morewords['3'][0] + '?')
    
    print('\n')
    print(morewords['1'][0].title(), e[0], e[1], morewords['2'][0], morewords['1'][2], morewords['3'][1] + '?')
    print(morewords['2'][4])
    print(morewords['1'][0].title(), e[0], e[1], morewords['2'][0], morewords['1'][2], morewords['3'][2] + '?')
    
    print('\n')
    f = ['All', 'you', 'gotta', 'do', 'is', 'choose']
    li2f = ' '.join(f)
    print(li2f[:-6])
    print(li2f[-6:])
    
    print('\n')
    print(f'by {w}')

fullpoem()
```
