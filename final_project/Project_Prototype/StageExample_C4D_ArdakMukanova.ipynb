```python
# Stage Example with cameras and light

# Python Custom Commands in Cinema 4D 
# Final Assignment, Ardak Mukanova

import c4d
from c4d import gui
from random import randint # not the default in c4d python

def colorLight(x, y):
    r = randint(x,y) / 256.0
    g = randint(x,y) / 256.0
    b = randint(x,y) / 256.0
    return c4d.Vector(r, g, b)


# Main function
def main():
""" this is a script for Cinema 4D to make simple stage with background, random lights and fixed cameras 
    the scene also has environment, sky and fog
    the colors of lights are randomized each script execute """
    
    def tool():
        return c4d.plugins.FindPlugin(doc.GetAction(), c4d.PLUGINTYPE_TOOL)

    def object():
        return doc.GetActiveObject()

    def tag():
        return doc.GetActiveTag()

    def renderdata():
        return doc.GetActiveRenderData()

    def prefs(id):
        return c4d.plugins.FindPlugin(id, c4d.PLUGINTYPE_PREFS)

    doc.StartUndo()

    #backstage
    ext = c4d.CallCommand(5116) # Extrude
    object()[c4d.EXTRUDEOBJECT_EXTRUSIONOFFSET] = 3000
    object()[c4d.ID_BASEOBJECT_REL_POSITION,c4d.VECTOR_Z] = -1500
    object()[c4d.ID_BASEOBJECT_REL_POSITION,c4d.VECTOR_Y] = 450
    object()[c4d.ID_BASEOBJECT_REL_POSITION,c4d.VECTOR_X] = 560

    arc = c4d.CallCommand(5182) # Arc
    object()[c4d.PRIM_ARC_RADIUS] = 1000
    object()[c4d.PRIM_ARC_START] = 0.105
    object()[c4d.PRIM_ARC_END] = 2.094
    object()[c4d.ID_BASEOBJECT_REL_ROTATION,c4d.VECTOR_Z] = -2.705


    arc = doc.GetActiveObject() # Look for the currently selected object
    if arc == None: return # if there is no object selected, quit
    ext = arc.GetNext() # Look for the object following the current object in the manager
    if ext == None: return # if there is no following object, quit

    arc.InsertUnder(ext)

    #cameras
    #cam 1
    cam1 = c4d.CallCommand(5103) # Camera
    c4d.CallCommand(55000, 2) # Use Camera
    object()[c4d.ID_BASEOBJECT_REL_POSITION,c4d.VECTOR_X] = 1000
    object()[c4d.ID_BASEOBJECT_REL_POSITION,c4d.VECTOR_Y] = 390
    object()[c4d.ID_BASEOBJECT_REL_POSITION,c4d.VECTOR_Z] = 460
    object()[c4d.ID_BASEOBJECT_REL_ROTATION,c4d.VECTOR_X] = 2.007
    object()[c4d.ID_BASEOBJECT_REL_ROTATION,c4d.VECTOR_Y] = -0.349
    object()[c4d.ID_BASEOBJECT_REL_ROTATION,c4d.VECTOR_Z] = 0

    c4d.CallCommand(5629) #protection for the camera
    c4d.CallCommand(100004788, 50062) # New Tag — protection

    #cam 2
    cam2 = c4d.CallCommand(5103) # Camera
    object()[c4d.ID_BASEOBJECT_REL_POSITION,c4d.VECTOR_X] = 980
    object()[c4d.ID_BASEOBJECT_REL_POSITION,c4d.VECTOR_Y] = 340
    object()[c4d.ID_BASEOBJECT_REL_POSITION,c4d.VECTOR_Z] = 0
    object()[c4d.ID_BASEOBJECT_REL_ROTATION,c4d.VECTOR_X] = 1.571
    object()[c4d.ID_BASEOBJECT_REL_ROTATION,c4d.VECTOR_Y] = -0.349
    object()[c4d.ID_BASEOBJECT_REL_ROTATION,c4d.VECTOR_Z] = 0

    c4d.CallCommand(5629) #protection for the camera
    c4d.CallCommand(100004788, 50062) # New Tag — protection

    #cam 3
    cam3 = c4d.CallCommand(5103) # Camera
    object()[c4d.ID_BASEOBJECT_REL_POSITION,c4d.VECTOR_X] = 380
    object()[c4d.ID_BASEOBJECT_REL_POSITION,c4d.VECTOR_Y] = 390
    object()[c4d.ID_BASEOBJECT_REL_POSITION,c4d.VECTOR_Z] = -370
    object()[c4d.ID_BASEOBJECT_REL_ROTATION,c4d.VECTOR_X] = 0.803
    object()[c4d.ID_BASEOBJECT_REL_ROTATION,c4d.VECTOR_Y] = -0.646
    object()[c4d.ID_BASEOBJECT_REL_ROTATION,c4d.VECTOR_Z] = 0

    c4d.CallCommand(5629) #protection for the camera
    c4d.CallCommand(100004788, 50062) # New Tag — protection


    #lights
    #light 1
    spt_light1 = c4d.CallCommand(202536) # Spot Light
    object()[c4d.ID_BASEOBJECT_REL_POSITION,c4d.VECTOR_X] = 1150
    object()[c4d.ID_BASEOBJECT_REL_POSITION,c4d.VECTOR_Y] = 1130
    object()[c4d.ID_BASEOBJECT_REL_POSITION,c4d.VECTOR_Z] = -2380
    object()[c4d.ID_BASEOBJECT_REL_ROTATION,c4d.VECTOR_X] = 0.436
    object()[c4d.ID_BASEOBJECT_REL_ROTATION,c4d.VECTOR_Y] = -0.436
    object()[c4d.ID_BASEOBJECT_REL_ROTATION,c4d.VECTOR_Z] = 0.175

    object()[c4d.LIGHT_BRIGHTNESS] = 0.8
    object()[c4d.LIGHT_SHADOWTYPE] = 1
    object()[c4d.LIGHT_DETAILS_OUTERANGLE] = 0.698 #outer light angle
    object()[c4d.LIGHT_COLOR] = colorLight(0, 255)


    #light 2
    spt_light2 = c4d.CallCommand(202536) # Spot Light
    object()[c4d.ID_BASEOBJECT_REL_POSITION,c4d.VECTOR_X] = 970
    object()[c4d.ID_BASEOBJECT_REL_POSITION,c4d.VECTOR_Y] = 840
    object()[c4d.ID_BASEOBJECT_REL_POSITION,c4d.VECTOR_Z] = 1600
    object()[c4d.ID_BASEOBJECT_REL_ROTATION,c4d.VECTOR_X] = 2.618
    object()[c4d.ID_BASEOBJECT_REL_ROTATION,c4d.VECTOR_Y] = -0.436
    object()[c4d.ID_BASEOBJECT_REL_ROTATION,c4d.VECTOR_Z] = 0.611

    object()[c4d.LIGHT_BRIGHTNESS] = 0.8
    object()[c4d.LIGHT_SHADOWTYPE] = 1
    object()[c4d.LIGHT_DETAILS_OUTERANGLE] = 0.698 #outer light angle
    object()[c4d.LIGHT_COLOR] = colorLight(0, 255)

    #envoronment
    sky = c4d.CallCommand(5105) # Sky

    env = c4d.CallCommand(5106) # Environment
    object()[c4d.ENVIRONMENT_FOGENABLE] = True
    object()[c4d.ENVIRONMENT_FOGSTRENGTH] = 0.5
    object()[c4d.ENVIRONMENT_FOGDISTANCE] = 30000
    object()[c4d.ENVIRONMENT_FOG] = colorLight(0, 255)

    doc.EndUndo()

# Execute main()
if __name__=='__main__':
    main()
    c4d.EventAdd()
```
