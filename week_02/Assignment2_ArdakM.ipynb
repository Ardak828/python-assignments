```python
def printpoem():
    a = "Chanson Dada"
    b = "Tristan Tzara"
    
    x = "dada"
    y = "heart"
    tuplelist = [(x, y)]
    
    print(f'{a.upper()}')
    print ('\n')
    print(f'this is the song of a {x}ist')
      
    for x,y in tuplelist: 
        print(f'who had {x} in his {y}')
        
    def beg(name='Unknown'):
        print(f'he {name}')
    beg('tore his motor apart')
    beg(f'had {x} in his {y}')
    
    print ('\n')
    print('the elevator lugged a king')
    beg('was a lumpy frail machine')
    beg('cut his right arm to the bone')
    print('sent it to the pope in rome')
  
    print ('\n')
    list = ['that’s why later', 'the elevator', 'had no more dada in its heart']
    for lines in list:
         print(lines)

    print ('\n')
    print('eat your chocolate')
    print('wash your brain')
    print(f'{x}')
    print(f'{x}')
    print('gulp some rain')
    
    print ('\n')
    print('this is the song of a bicyclist')
    print(f'who loved {x} from the start')
    print(f'she therefore was a {x}ist')
    
    for x,y in tuplelist: 
        print(f'like all with {x} in their {y}')
    
    print ('\n')
    print('but her husband on new year’s day')
    print('learned everything & in a crisis')
    print('sent to the vatican right away')
    print('their two bodies in two suitcases')
    
    print ('\n')
    list = ['nor the bicyclist', 'nor the man', 'was ever happy or sad again']
    for lines in list:
         print(lines)
    
    print ('\n')
    print('drink some bird’s milk')
    print('wash your sweets')
    print(f'{x}')
    print(f'{x}')
    print('eat your meat')
    
    print ('\n')
    print(f'a {x}ist poem by {b}'.title())

printpoem()
```
